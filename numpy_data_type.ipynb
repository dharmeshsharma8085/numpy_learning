binaray files(.npy, .np)

```python
arr=np.array([1,2,3,4])
np.save("array.npy",arr) # we can't read binary file with noraml txt reader
```

```python
arr=np.load("array.npy")
print(arr)
```

```python
#save multiple file
# not possible in txt file  only possible in binary file
a=np.array([1,2,3])
b=np.array([5,6,7])
np.savez("multi_array.npz", arr1=a , arr2=b)
#use savez to save multi file and npz
```

```python
arr=np.load("multi_array.npz")
print(arr["arr1"])
print(arr["arr2"])
```
