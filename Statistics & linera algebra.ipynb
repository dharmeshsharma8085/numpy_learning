Linear algebra Using numpy


```python
A = np.arange(1,10).reshape(3,3)
B = ri(1,10,9).reshape(3,3)
print(A)
print(B)
```

```python
# Dot prodocut of A and B ( for 2D array it is eqivalent to matrix multiplication)
np.dot(A,B)
```

```python
#1d
A=np.arange(1,6)
B=ri(1,10,5)
print(A)
print(B)
```

```python
np.dot(A,B)
np.inner(A,B) # same as dot for 1d
```

```python
#outer 2d
np.outer(A,B)
```

Matrix Transpose and Matrix inverse

```python
A = ri(1,10,9).reshape(3,3)
A
```

```python
#transpose
np.transpose(A)
```

```python
# matrix multiplication of A and A transpose
np.dot(A,A.T)
```

Finding Trace of Matrix

```python
A = ri(1,10,16).reshape(4,4)
A
```

```python
#Matrix Trace
np.trace(A)
# sum of main diagonal
```

```python
# Matrix Trace With ofset +1 (upper traingle)
np.trace(A , offset=1)
```

```python
# Matrix Trace With ofset -1 (lower traingle)

np.trace(A ,offset=-1)
```
